```python
import sys
from pathlib import Path
sys.path.append(str(Path('..').resolve()))

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from src.loan_generator import loan_generator

df_clients = pd.DataFrame({'id': np.arange(1, 50_001)})

df_loans = loan_generator(df_clients)

sns.lineplot(data=df_loans, x='loan_date', y='loan_amount')
plt.show()
```

```python
df_loans['loan_date'] = pd.to_datetime(df_loans['loan_date'])

df_loans['year'] = df_loans['loan_date'].dt.year

df_year = (
    df_loans
    .groupby('year')
    .size()
    .reset_index(name='qtd_loans')
)

sns.lineplot(data=df_year, x='year', y='qtd_loans')
plt.show()
```
